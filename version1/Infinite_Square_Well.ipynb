```python
"""
Landon Buell
The Quantum Game
Infinite Square Well
19 December 2019
"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4,12,0.01)
L = 8
V = np.zeros(len(x))

pts0 = np.where((x<0))
pts1 = np.where((x>L))
V[pts0] += 100
V[pts1] += 100
```

```python
plt.figure(figsize=(20,8))
plt.title("Infinite Square Well",size=40,weight='bold')
plt.xlabel("X-Axis Position",size=20,weight='bold')
plt.ylabel("Energy Value",size=40,weight='bold')

plt.xlim(-4,+12)
plt.ylim(-10,+110)

plt.xticks(np.arange(-4,13,2),
           ['$-L/2$','$-L/4$','$0$','$L/4$','$L/2$','$3L/4$','$L$','$5L/4$','$3L/2$'])
plt.yticks([0,100],['$0$','$+\infty$'])

plt.plot(x,V,color='red',label='V(x)')
plt.hlines(0,-4,+12,color='black')
plt.vlines(0,-10,+110)

plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('ISW.png')
plt.show()
```
