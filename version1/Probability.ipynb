```python
""" Create a distribution """

years = np.arange(10,31,1)
amnts = np.random.randint(0,6,size=len(years))
print(years)
print(amnts)

plt.figure(figsize=(20,8))
plt.title("Standard Probability Exercise",size=40,weight='bold')
plt.xlabel("Years of Ages",size=20,weight='bold')
plt.ylabel("Number of Occurances",size=20,weight='bold')

plt.hist(amnts,bins=years)

plt.grid()
plt.tight_layout()
plt.show()
```
