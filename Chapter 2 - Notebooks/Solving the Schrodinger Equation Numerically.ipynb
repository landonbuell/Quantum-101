```python
import numpy as np

# Establish number of steps, space,and step size
N = 256
x = np.linspace(0,2*np.pi,N)
dx = x[1] - x[0]
```

```python
# Build First Derivative Matrix

def FirstDerivativeMatrix(Npts,stepsize):
    """ Build N x N Matrix Representation of 1st-Derivative Operator """
    mainDiagonal = np.diag(np.ones(Npts),0) * -1
    upperDiagonal = np.diag(np.ones(Npts-1),1)
    matrix = (mainDiagonal + upperDiagonal)/stepsize
    return matrix
    
```

```python
# Function to use as f(x)
y = np.sin(x)

# Compute known 1st-Derivative
dy_exact = np.cos(x)

# Compute Numerical 1st-Derivative
d1Matrix = FirstDerivativeMatrix(N,dx)
dy = np.matmul(d1Matrix,np.transpose(y))
```

```python
# Plot Results 

import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.title("Numerical First Derivative Comparison",fontsize=50,fontweight='bold')
plt.xlabel("Input Space",fontsize=40,weight='bold')
plt.ylabel("Output Space",fontsize=40,weight='bold')

plt.plot(x,y,color='blue',linestyle='-',linewidth=4,label="$\sin(x)$")
plt.plot(x,dy_exact,color='green',linestyle='-.',linewidth=4,label="$\cos(x)$")
plt.plot(x,dy,color='red',linestyle='--',linewidth=4,label="$\\frac{d}{dx}[\sin(x)]$")

plt.hlines(0,0,2*np.pi,color='black')
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.legend(fontsize=40)
plt.grid()
plt.tight_layout()

plt.show()
```
